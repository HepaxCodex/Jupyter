```python
import urllib, json

url = "https://static01.nyt.com/elections-assets/2020/data/api/2020-11-03/votes-remaining-page/national/president.json"
response = urllib.request.urlopen(url)
electionData = json.loads(response.read())
```

```python
#for i, race in enumerate(electionData["data"]["races"]):
#    print(i, race["race_id"])

keyRacesIndexs = [3, 10, 27, 33]
for i, index in enumerate(keyRacesIndexs):
    print(electionData["data"]["races"][index]["race_id"])
    print('\t', electionData["data"]["races"][index]["expectations_text"])
```
