# 2020 Election Tool from NYT


Based On https://github.com/alex/nyt-2020-election-scraper

```python
# First lets get the data from the JSON Site... 
#It may be helpful to copy-paste the link into a Browser.  
# Firefox provides a nice interface to view the data and look around to guide your coding.

import urllib, json

url = "https://static01.nyt.com/elections-assets/2020/data/api/2020-11-03/votes-remaining-page/national/president.json"
response = urllib.request.urlopen(url)
electionData = json.loads(response.read())["data"]["races"] # Trim the top layers data and races 

keyRacesIndexs = [3, 10, 27, 33, 38]
```

## Plots

```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from IPython.display import display
import datetime
from bqplot import pyplot as plt
import numpy as np
import pandas as pd

@interact(
    x=[(electionData[x]["race_id"], x) for x in keyRacesIndexs],
    y=['Vote Shares', 'Votes']
)
def makePlot(x,y):
    # Setup the data frame
    df = pd.DataFrame(electionData[x]["timeseries"])
    # Reformat the timestamps into numpy timestamps
    timestamps = [np.datetime64(x) for x in df["timestamp"]]
    # Generate the plot
    plt.figure(title=electionData[x]["race_id"])
    if(y == 'Vote Shares'):
        plt.plot(timestamps[1:], [x["vote_shares"]["bidenj"]*100 for x in electionData[x]["timeseries"]][1:], '-b')
        plt.plot(timestamps[1:], [x["vote_shares"]["trumpd"]*100 for x in electionData[x]["timeseries"]][1:], '-r')
        plt.ylabel('Vote Share %')
    else:
        plt.plot(timestamps[1:], [x["votes"] for x in electionData[x]["timeseries"]][1:])
        plt.ylabel('Total Votes')
        
    plt.show()
```
