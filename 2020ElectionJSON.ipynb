## First lets get the data from the JSON Site... 
It may be helpful to copy-paste the link into a Browser.  Firefox provides a nice interface to view the data and look around to guide your coding.

```python
import urllib, json

url = "https://static01.nyt.com/elections-assets/2020/data/api/2020-11-03/votes-remaining-page/national/president.json"
response = urllib.request.urlopen(url)
electionData = json.loads(response.read())
```

## Now let print the "Expectations Text" for some of the key races
The first (commented out) part prints out just the index and its associated race_id.
The next part selects indexes we care about (Arizona, Georgia, North Carolina, Nevada, Pensylvenia)

```python
#for i, race in enumerate(electionData["data"]["races"]):
#    print(i, race["race_id"])

keyRacesIndexs = [3, 10, 27, 33, 38]
for i, index in enumerate(keyRacesIndexs):
    print(electionData["data"]["races"][index]["race_id"])
    print('\t', electionData["data"]["races"][index]["expectations_text"])
```

```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from IPython.display import display
import datetime

@interact(x=[(electionData["data"]["races"][x]["race_id"], x) for x in keyRacesIndexs])
def makePlot(x):
    df = pd.DataFrame(electionData["data"]["races"][x]["timeseries"])
    display(type(df["timestamp"][1]))
    display(df["votes"])
    plt.figure(title=electionData["data"]["races"][x]["race_id"])
    plt.plot(df["votes"])
    plt.show()
```
